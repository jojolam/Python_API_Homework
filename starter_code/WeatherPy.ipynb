```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Define URL
url = "http://api.openweathermap.org/data/2.5/weather?"

#Create empty list to append data to
ls_city_name = []
ls_max_temp = []
ls_wind_speed = []
ls_humidity = []
ls_lat = []
ls_cloudiness = []
ls_lng = []
ls_country = []
ls_date = []

record = 1

#print statement
print(f"Beginning Data Retrieval")
print(f"-------------------------------")

#looping throuh all cities and print statement
for city in cities :
    try: 
         query_url = f"{url}appid={api_key}&q={city}"
         response = requests.get(query_url).json()
    
         ls_city_name.append(response["name"])
         ls_cloudiness.append(response["clouds"]["all"])
         ls_max_temp.append(response["main"]["temp_max"])
         ls_wind_speed.append(response["wind"]["speed"])
         ls_humidity.append(response["main"]["humidity"])
         ls_lat.append(response["coord"]["lat"])
         ls_lng.append(response["coord"]["lon"])
         ls_country.append(response["sys"]["country"])
         ls_city_record = response["name"]
         ls_date.append(response["dt"])
    
         print (f"Processing Record{record}|{ls_city_record}")
         print(f"{url}&q={city}")
    
         record= record + 1
     
         time.sleep(0.1)
    
    except:
        print("City not found. Skipping...")
    continue
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#create weather dictionary
weather_dict = {
    "City": ls_city_name,
    "Cloudiness":ls_cloudiness, 
    "Country":ls_country,
    "Date":ls_date, 
    "Humidity": ls_humidity,
    "Lat":ls_lat, 
    "Lng":ls_lng, 
    "Max Temp": ls_max_temp,
    "Wind Speed":ls_wind_speed
}

#create dataframe
weather_df = pd.DataFrame(weather_dict)

#count of the values in dataframe
weather_df.count()
```

```python
#save to CSV
weather_df.to_csv('Output_CSV/weather_data.csv')
#print the head of dataframe
weather_df.head()
```
